# SEC DATA ANALYSIS

```python
import pandas as pd
```

```python
df = pd.read_csv('10-K Filings.csv')
```

```python
df
```

## Trend Analysis

```python
df['Revenue Growth (%)'] = df.groupby(['COMPANIES'])['TOTAL REVENUE'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['COMPANIES'])['NET INCOME'].pct_change() * 100
```

```python
df
```

* The first year for each company lacks a previous year for comparison

```python
avg_growth = df.groupby('COMPANIES')[['Revenue Growth (%)', 'Net Income Growth (%)']].mean()
yearly_revenue = df.groupby('YEARS')['TOTAL REVENUE'].sum()
df
```

```python
avg_growth
```

```python
yearly_revenue
```

```python
df['Assets Growth (%)'] = df.groupby(['COMPANIES'])['TOTAL ASSETS'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby(['COMPANIES'])['TOTAL LIABILITIES'].pct_change() * 100
df['Cash Flow(OS) Growth (%)'] = df.groupby(['COMPANIES'])['CASH FLOW(OS)'].pct_change() * 100
df['Profit Margin (%)'] = (df['NET INCOME'] / df['TOTAL REVENUE']) * 100
df['Debt-to-Assets Ratio'] = (df['TOTAL LIABILITIES'] / df['TOTAL ASSETS']) 
df
```

```python
df.fillna(0, inplace=True)
df
```
